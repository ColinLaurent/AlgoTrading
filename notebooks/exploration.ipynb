```python
import requests
```

```python
BASE_URL = "https://api.binance.com"
SYMBOL = "BTCUSDT"
```

```python
r = requests.get(f"{BASE_URL}/api/v3/ticker/price", params={"symbol": SYMBOL}, timeout=10)
r.raise_for_status()
data = r.json()
```

```python
print(f"{data['symbol']} price = {data['price']}")
```
